```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("train.csv")

print(df.shape)
df.head()

df.info()

print(df.isnull().sum())

print(df.describe())

df['ESTU_PRGM_ACADEMICO'].value_counts().head(10).plot(kind='barh', figsize=(10,6), title="Top 10 Programas Acadmicos")
plt.xlabel("Cantidad de Estudiantes")
plt.show()

df['ESTU_PRGM_DEPARTAMENTO'].value_counts().plot(kind='bar', figsize=(12,6), title="Estudiantes por Departamento")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.show()

df['RENDIMIENTO_GLOBAL'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Distribucin de RENDIMIENTO_GLOBAL', figsize=(6,6))
plt.ylabel('')
plt.show()
```
